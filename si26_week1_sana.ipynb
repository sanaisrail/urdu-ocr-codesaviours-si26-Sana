1. What is OCR (Optical Character Recognition)?

Answer:  Optical Character Recognition (OCR) is a technology that helps computers read text from images, scanned documents, or photos. It converts the text into a digital format that can be edited, copied, or searched. OCR is very useful because it saves time and reduces the need to type everything manually.

2. Why is Urdu OCR harder than English OCR?

Answer: Urdu OCR is more difficult than English OCR because Urdu is written from right to left, and many letters change their shape depending on where they appear in a word. Some Urdu letters also look very similar, with only small differences like dots. Different handwriting styles and unclear images make it even harder for a computer to recognize the text correctly.

3. What are two real-world situations where Urdu OCR would be useful?

Answer:Urdu OCR can help digitize old books, newspapers, and historical documents so that they can be preserved and searched easily in the future. It is also useful for businesses and government offices to convert printed Urdu forms, records, and official documents into editable digital text, reducing manual work and improving efficiency. These applications save time, improve accuracy, and make important information easier to access and manage.
