1. What is OCR (Optical Character Recognition)?

Answer:  Optical Character Recognition (OCR) is a technology that helps computers read text from images, scanned documents, or photos. It converts the text into a digital format that can be edited, copied, or searched. OCR is very useful because it saves time and reduces the need to type everything manually.

2. Why is Urdu OCR harder than English OCR?

Answer: Urdu OCR is more difficult than English OCR because Urdu is written from right to left, and many letters change their shape depending on where they appear in a word. Some Urdu letters also look very similar, with only small differences like dots. Different handwriting styles and unclear images make it even harder for a computer to recognize the text correctly.

3. What are two real-world situations where Urdu OCR would be useful?

Answer:Urdu OCR can help digitize old books, newspapers, and historical documents so that they can be preserved and searched easily in the future. It is also useful for businesses and government offices to convert printed Urdu forms, records, and official documents into editable digital text, reducing manual work and improving efficiency. These applications save time, improve accuracy, and make important information easier to access and manage.

```python
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
import os

urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے",
    "علم روشنی ہے",
    "وقت بہت قیمتی ہے",
    "محنت کا پھل میٹھا ہوتا ہے",
    "ہم سب پاکستانی ہیں",
    "صاف ماحول صحت کی ضمانت ہے",
    "درخت لگائیں ماحول بچائیں",
    "کتاب بہترین دوست ہے",
    "ہر انسان برابر ہے",
    "سچ ہمیشہ کامیاب ہوتا ہے",
    "پاکستان ایک خوبصورت ملک ہے",
    "اردو ہماری قومی زبان ہے",
    "پڑھائی کامیابی کی پہلی سیڑھی ہے",
    "ہمیں دوسروں کی مدد کرنی چاہیے",
    "دیانت داری بہترین پالیسی ہے",
    "اتحاد ایمان اور نظم"
]

os.makedirs("data/synthetic", exist_ok=True)

font = ImageFont.truetype(
    "/usr/share/fonts/truetype/noto/NotoNaskhArabic-Regular.ttf",
    32
)

for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)

    img = Image.new("RGB", (700, 120), "white")
    draw = ImageDraw.Draw(img)

    bbox = draw.textbbox((0, 0), bidi_text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]

    x = 700 - text_width - 20
    y = (120 - text_height) // 2

    draw.text((x, y), bidi_text, fill="black", font=font)

    img.save(f"data/synthetic/urdu_{i+1}.png")

print(" 20 Synthetic Images Generated Successfully!")
```

```python
import os
import csv

labels = []

folders = [
    "data/raw/newspaper",
    "data/raw/books",
    "data/raw/signboards",
    "data/raw/synthetic",
    "data/raw/other"
]

for folder in folders:
    for file in os.listdir(folder):
        if file.lower().endswith(".png"):
            image_path = os.path.join(folder, file)

            base = os.path.splitext(image_path)[0]
            txt_path = base + ".txt"

            if os.path.exists(txt_path):
                with open(txt_path, "r", encoding="utf-8") as f:
                    text = f.read().strip()
            else:
                text = ""

            labels.append({
                "image": image_path,
                "text": text
            })

with open("data/labels.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["image", "text"])
    writer.writeheader()
    writer.writerows(labels)

print(f"Labels file created with {len(labels)} entries")
```
